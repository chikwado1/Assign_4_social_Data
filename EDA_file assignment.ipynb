```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data
salary = [2, 4, 6, 8, 10]
rent_expenses = [3, 5, 7, 9, 11]
transport = [4, 6, 8, 10, 12]
food_expenses = [5, 7, 9, 11, 13]

# Calculate summary statistics for each variable
variables = {'Salary': salary, 'Rent Expenses': rent_expenses, 'Transport': transport, 'Food Expenses': food_expenses}
statistics = {var: {'Mean': np.mean(data), 'Median': np.median(data), 'Std. Dev.': np.std(data),
                    'Sum': np.sum(data), 'Min': np.min(data), 'Max': np.max(data)} for var, data in variables.items()}

# Create a figure and axis object
fig, ax = plt.subplots(figsize=(10, 6))

# Plot histograms for all four variables
for var, data in variables.items():
    ax.hist(data, alpha=0.5, label=var, bins=5)

    # Annotate summary statistics
    ax.annotate(f"Mean: {statistics[var]['Mean']:.2f}\nMedian: {statistics[var]['Median']:.2f}\nStd. Dev.: {statistics[var]['Std. Dev.']:.2f}\nSum: {statistics[var]['Sum']}\nMin: {statistics[var]['Min']}\nMax: {statistics[var]['Max']}",
                xy=(statistics[var]['Mean'], 0), xycoords='data',
                xytext=(5, 10), textcoords='offset points',
                arrowprops=dict(facecolor='black', arrowstyle='->', connectionstyle='arc3,rad=0.3'))

# Add labels and title
ax.set_xlabel('Value')
ax.set_ylabel('Frequency')
ax.set_title('Histogram of Four Variables')
ax.legend()

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data
ages = [25, 30, 35, 40, 45]
salary = [2, 4, 6, 8, 10]
rent_expenses = [3, 5, 7, 9, 11]
transport = [4, 6, 8, 10, 12]
food_expenses = [5, 7, 9, 11, 13]

# Calculate summary statistics for each variable
variables = {'Salary': salary, 'Rent Expenses': rent_expenses, 'Transport': transport, 'Food Expenses': food_expenses}
statistics = {var: {'Mean': np.mean(data), 'Median': np.median(data), 'Std. Dev.': np.std(data),
                    'Sum': np.sum(data), 'Min': np.min(data), 'Max': np.max(data)} for var, data in variables.items()}

# Create a figure and axis object
fig, ax = plt.subplots(2, 2, figsize=(12, 8))

# Plot scatter plots for each variable against ages
for i, (var, data) in enumerate(variables.items()):
    row = i // 2
    col = i % 2

    ax[row, col].scatter(ages, data, label=var)
    ax[row, col].set_xlabel('Ages')
    ax[row, col].set_ylabel(var)

    # Annotate summary statistics
    ax[row, col].annotate(f"Mean: {statistics[var]['Mean']:.2f}\nMedian: {statistics[var]['Median']:.2f}\nStd. Dev.: {statistics[var]['Std. Dev.']:.2f}\nSum: {statistics[var]['Sum']}\nMin: {statistics[var]['Min']}\nMax: {statistics[var]['Max']}",
                          xy=(ages[-1], data[-1]), xycoords='data',
                          xytext=(-50, -30), textcoords='offset points',
                          arrowprops=dict(facecolor='black', arrowstyle='->', connectionstyle='arc3,rad=0.3'))

# Adjust layout
plt.tight_layout()

# Display the plot
plt.show()
```

```python
import numpy as np
import plotly.express as px

# Sample data
ages = [25, 30, 35, 40, 45]
salary = [2, 4, 6, 8, 10]
rent_expenses = [3, 5, 7, 9, 11]
transport = [4, 6, 8, 10, 12]
food_expenses = [5, 7, 9, 11, 13]

# Calculate summary statistics for each variable
variables = {'Salary': salary, 'Rent Expenses': rent_expenses, 'Transport': transport, 'Food Expenses': food_expenses}
statistics = {var: {'Mean': np.mean(data), 'Median': np.median(data), 'Std. Dev.': np.std(data),
                    'Sum': np.sum(data), 'Min': np.min(data), 'Max': np.max(data)} for var, data in variables.items()}

# Create scatter plots
fig = px.scatter(y=salary, labels={'y': 'Salary'}, title='Scatter Plot: Ages vs Salary')

# Add scatter traces for each variable
for var, data in variables.items():
    fig.add_scatter(x=ages, y=data, mode='markers', name=var)

    # Add annotations for summary statistics
    fig.add_annotation(x=ages[-1], y=data[-1],
                       text=f"Mean: {statistics[var]['Mean']:.2f}<br>Median: {statistics[var]['Median']:.2f}<br>Std. Dev.: {statistics[var]['Std. Dev.']:.2f}<br>Sum: {statistics[var]['Sum']}<br>Min: {statistics[var]['Min']}<br>Max: {statistics[var]['Max']}",
                       showarrow=True, arrowhead=1, arrowcolor='black', ax=-50, ay=-30)

# Show the plot
fig.show()
```
